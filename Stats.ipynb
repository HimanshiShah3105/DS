```python
from random import randint, randrange
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = [randint(10, 20) for i in range(20)]
print(x)
```

```python
def mean(arr):
    sum_ = 0
    for i in arr:
        sum_ += i
    m_ = sum_ / len(arr)
    return m_
```

```python
mean(x)
```

```python
def median(arr):
    l = sorted(arr)
    if len(arr) % 2 == 0:
        m1 = l[len(l)//2]
        m2 = l[len(l)//2-1]
        median = (m1+m2)/2
    else:
        median = l[(len(l)+1)//2]
    return median
    
```

```python
median(x)
```

```python
font = {"fontsize":20, "color":"red"}
m = mean(x)
med = median(x)
plt.plot(x, 'o')
plt.plot([0, 20], [m, m], 'k--')
plt.plot([0, 20], [med, med], 'r--')
for i in range(len(x)):
    plt.plot([i, i], [m, x[i]], 'b-')
plt.xlabel("INDEX", fontdict=font)
plt.ylabel("X VALUE", fontdict=font)
plt.xticks(range(20))
plt.show()
```
